```python
#Above table, find percentage of cases and vaccinations
```

```python
#filter out rows with Null
vaccinations=vaccinations.dropna(axis=0)
```

```python
#create a new column and populate with total case percentages
vaccinations['total_cases_percentage']=(vaccinations['total_cases'].div(vaccinations.population))*100
#create a new column with total vaccination percentages
vaccinations['total_vaccinations_percentage']=(vaccinations['total_vaccinations'].div(vaccinations.population))*100
```

```python
vaccinations.head()
```

```python
#Finding averages and grouping by location
VaxCaseAverages=vaccinations.groupby('location')['total_cases'].agg(['mean'])
VaxCaseAverages.rename(columns={'mean': 'Total Case Mean'}, inplace=True)
VaxVaxAverages=vaccinations.groupby('location')['total_vaccinations'].agg(['mean'])
VaxVaxAverages.rename(columns={'mean': 'Total Vaccines Mean'}, inplace=True)
VaxPopAverages=vaccinations.groupby('location')['population'].agg(['mean'])
VaxPopAverages.rename(columns={'mean': 'Population Mean'}, inplace=True)
VaxCasePercentAverages=vaccinations.groupby('location')['total_cases_percentage'].agg(['mean'])
VaxCasePercentAverages.rename(columns={'mean': 'Total Cases Percentage Mean', 'mean_y': 'Cases Mean'}, inplace=True)
VaxCaseVaxPercentAverages=vaccinations.groupby('location')['total_vaccinations_percentage'].agg(['mean'])
VaxCaseVaxPercentAverages.rename(columns={'mean': 'Total Vaccines Percentage Mean'}, inplace=True)
```

```python
#Combine the tables
TotalAverages=VaxCaseAverages.merge(VaxVaxAverages, on="location")
TotalPopAverages=TotalAverages.merge(VaxPopAverages, on="location")
TotalPopCasePercentAverages=TotalPopAverages.merge(VaxCasePercentAverages, on="location")
VaxCasePop=TotalPopCasePercentAverages.merge(VaxCaseVaxPercentAverages, on="location")
VaxCasePop.head()
```

```python
#calculate correlation between the percentage columns we created
vacCorr=VaxCasePop.corr()
```

```python
#create a heatmap of the correlation
plt.figure()
sns.heatmap(vacCorr, annot=True, linewidth=0.5, cmap="viridis")
plt.title("Correlation between Cases and Vaccinations")
plt.savefig("Correlation between Cases and Vaccinations Heatmap")
plt.show()
```
